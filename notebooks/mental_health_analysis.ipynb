# 📊  Univariate Analysis

In this step, we analyze each feature individually to understand its distribution, spread, and patterns.

This helps us:
- Identify common user behavior
- Detect outliers
- Understand how values are distributed

## 📌 Distribution of Screen Time

```python
plt.hist(df['daily_screen_time_min'])
plt.title("Distribution of Screen Time")
plt.xlabel("Daily Screen time")
plt.ylabel("Frequency")
plt.savefig("../images/Screen_distribution.png")  
plt.show()
```

### 📊 Screen Time Analysis

- The distribution is approximately normal (bell-shaped), indicating most users have moderate screen time.
- The screen time ranges from around 180 to 600 minutes per day.
- Most users fall within the 300–400 minutes range, showing high daily usage.
- A few users have very low (<200) and very high (>500) screen time, indicating outliers.

💡 Insight:
Most users spend a significant amount of time on screens daily, suggesting that high screen usage is a common behavior.

## 📌 Distribution of Sleep Hours

```python
plt.hist(df['sleep_hours'])
plt.title("Distribution of Sleep Hours")
plt.xlabel("Sleep Hours")
plt.ylabel("Frequency")
plt.savefig("../images/Sleep_distribution.png")
plt.show()
```

### 😴 Sleep Hours Analysis

- The distribution is approximately normal, indicating balanced sleep patterns among users.
- Sleep duration ranges from 3 to 10 hours.
- Most users fall in the 5–8 hours range, with a peak around 6–7 hours.
- There are a few users with very low (<4 hours) and high (>9 hours) sleep durations.

💡 Insight:
Most users maintain moderate sleep, but a portion of users with lower sleep hours may be at risk of negative mental health effects.

## 📌 Distribution of Notification Count

```python
plt.hist(df['notification_count'])
plt.title("Distribution of Notification Count")
plt.xlabel("Notification Count")
plt.ylabel("Frequency")
plt.savefig("../images/notification_distribution.png")
plt.show()
```

### 🔔 Notification Count Analysis

- The distribution is approximately normal but slightly right-skewed.
- Notification count ranges from around 20 to 160.
- Most users fall in the 60–100 range, indicating moderate notification exposure.
- A few users receive very high notifications (>130), suggesting possible outliers.

💡 Insight:
Most users experience a moderate level of notifications, but high notification counts for some users may lead to frequent interruptions and reduced focus.

## 📌 Distribution of Anxiety Level

```python
plt.hist(df['anxiety_level'])
plt.title("Distribution of Anxiety Level")
plt.xlabel("Anxiety Level")
plt.ylabel("Frequency")
plt.savefig("../images/anxiety_distribution.png")
plt.show()
          
```

### 😰 Anxiety Level Analysis

- The distribution is left-skewed, indicating that most values are concentrated at the higher end.
- Anxiety levels range from approximately 1 to 10.
- Most users fall within the 8–10 range, showing high anxiety levels.
- A smaller number of users have low anxiety levels (1–4).

💡 Insight:
A large proportion of users exhibit high anxiety levels, suggesting that many users may be experiencing significant stress or mental pressure.
