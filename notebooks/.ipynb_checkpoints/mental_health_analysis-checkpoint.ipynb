```python
plt.hist(df['daily_screen_time_min'])
plt.title("Distribution of Screen Time")
plt.xlabel("Daily Screen time")
plt.ylabel("Frequency")
plt.savefig("../images/Distribution_of_screen_time.png")  
plt.show()
```
